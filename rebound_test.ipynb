```python
import rebound
import numpy as np
```

```python
sim = rebound.Simulation()
sim.add(m=1)
sim.add(m=1e-3, a=1.,e=0.1)
sim.add(a=1.4,e=0.1)

sim.integrate(100.)
```

```python
for p in sim.particles:
    print(p.x, p.y, p.z)
for o in sim.orbits(): 
    print(o)
```

```python
rebound.OrbitPlot(sim,unitlabel="(AU)")
```

```python
sim2 = rebound.Simulation()
sim2.add(m=1)
sim2.add(a=1.5,e=0.6)
sim2.add(x=1,y=1.5, vx=0.15,vy=-0.3)
sim2.add(a=2,e=0.4,omega=np.pi,inc=np.radians(15),Omega = np.radians(45))
sim2.add(a=0.5,e=0.8,omega=np.radians(120))

sim2.status()
rebound.OrbitPlot(sim2)
```

```python
simSolarSys = rebound.Simulation()
rebound.data.add_solar_system(simSolarSys)
rebound.data.add_outer_solar_system(simSolarSys)

rebound.OrbitPlotSet(simSolarSys)
```

```python
rebound.OrbitPlotSet(sim2,slices=True)
```

```python
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

animate = rebound.Simulation()
animate.add(m=1)
animate.add(a=1)

op = rebound.OrbitPlot(animate)
fig = op.fig

for i in range(200):
    animate.integrate(animate.t + 0.1)
    op.update()
    clear_output(wait=True)
    display(fig)
    plt.close(fig)
```
